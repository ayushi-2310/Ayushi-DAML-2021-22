# How to handle missing values using pandas? 

```python
v1=pd.Series([1,np.nan,2,None])
v1
```

```python
data={'student':['Jack','Mike','Rohan','Zubair','Mihir',np.nan], 'year':[1,np.nan,3,1,2,5], 'marks':[9.8,6.1,np.nan,9.7,4.3,6.8]}
df=pd.DataFrame(data)
df
```

```python
df.isnull()
```

```python
fd.isnull()
```

```python
df.isnull().sum()
```

```python
df.notnull().sum()
```

```python
df.dropna()
```

```python
fd.dropna()
```

```python
df.dropna(how='all')
```

```python
df.dropna(how='all',axis=1)
```

```python
df.marks.mean()
```

To fill missing value in datset

```python
df.fillna(0)
```

```python
df.fillna(50)
```

```python
df.fillna(method='ffill')       #forward
```

```python
df.fillna(method='bfill')  #backward
```

```python
df.fillna({'marks':7.34})
```
