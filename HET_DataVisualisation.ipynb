#ML
In linear regression, X is 2D array and Y is 1D array

```python
df=pd.read_csv('/content/House_Prices.csv')
df.head(2)
```

```python
X=df.iloc[:,1].values.reshape(-1,1)
y=df.iloc[:,2].values
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=21)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pre=lr.predict(X_test)
y_pre
```

```python
y_pre-y_test
```

```python
lr.score(X_test,y_test)
```

```python
sns.scatterplot(x='Price',y='SqFt',data=df)
```

```python
plt.plot(df.iloc[:,0],lr.predict(X),color="red")
```
